```python
#Avoiding overlapping:Rotation
plt.figure(figsize=(12,8))
plt.bar(x_bar,y_bar)
plt.xticks(rotation=90,fontsize=10)
```

```python
plt.figure(figsize=(12,8))
plt.bar(x_bar,y_bar,color="orange")#color change
plt.xticks(rotation=90,fontsize=10)
#labelling axids
plt.xlabel("Genre",fontsize=12)
plt.ylabel("Count",fontsize=12)
#Title
plt.title("Top Genre Analysis",fontsize=20)
plt.show()
```

#TOP 5 PUBLISHERS

```python
top_publisher=df["Publisher"].value_counts().index[:5]
print(top_publisher)
```

#TOP 5 GENRE

```python
top_genre=df["Genre"].value_counts().index[:5]
print(top_genre)
```

#TOP 5 PLATFORM

```python
top_platform=df["Platform"].value_counts().index[:5]
print(top_platform)
```

```python
top_5_data=df.loc[(df["Publisher"].isin(top_publisher)&(df["Platform"].isin(top_platform))&(df["Genre"].isin(top_genre)))]
print(top_5_data)
```

```python
sns.countplot(data=top_5_data,x="Publisher",hue="Platform")
plt.xticks(rotation=45)   # Rotate x-axis labels by 45 degrees
plt.show()
```

```python
#Stacked plot
df_stacked_plot=pd.crosstab(index=top_5_data["Publisher"],columns=top_5_data["Platform"])
print(df_stacked_plot)
```

```python
df_stacked_plot.plot(kind="bar",stacked=True)
plt.show()
```

#When we want to study about statsical analysis of each category in the data :Box plot can be used

```python
#finding outliers in each category
sns.boxplot(x="Publisher",y="Global_Sales",data=top_5_data)
plt.xticks(rotation=90,fontsize=20)
plt.show()
```
